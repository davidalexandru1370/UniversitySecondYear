```python
import torch
import torch.nn as nn
from collections import OrderedDict
from datetime import datetime

data_in = torch.tensor([[0, 0],[1, 0],[0, 1],[1, 1]], dtype=torch.float32)
data_out = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]], dtype=torch.float32)

print("data in = ", data_in)
print("data out = ", data_out)

model = nn.Sequential(OrderedDict([
    ('hidden', nn.Linear(2,4)),
    ('sigmoid', nn.Sigmoid()),
    ('output', nn.Linear(4,2)),
    ('sigmoid', nn.Sigmoid())
]))

print("model = ", model)


criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.25)
start_time = datetime.now()
stop_time = datetime.now()

for epoch in range(10000):
    optimizer.zero_grad()
    output = model(data_in)
    loss = criterion(output, data_out)
    loss.backward()
    optimizer.step()
    predicted_list = torch.round(output)
    correct = (predicted_list == data_out).sum()
    if correct == 1.0:
        stop_time = datetime.now()
    if (epoch + 1) % 1000 == 0:
        print(f'Epoch {epoch+1}/{10000} with loss =  {loss.item():.4f}')

        

#print(output)
predicted_list = torch.round(output)
#print(predicted_list == data_out)
correct = (predicted_list == data_out).sum()
accuracy = correct / (len(data_out) * 2)
print("Accuracy is = ", accuracy)

if accuracy == 1.0:
    start_time = datetime.strptime(start_time.strftime("%H:%M:%S:%f"),"%H:%M:%S:%f")
    stop_time = datetime.strptime(stop_time.strftime("%H:%M:%S:%f"),"%H:%M:%S:%f")
    print("Total time = ", stop_time - start_time)
    
print("Weights:")
index:int = 0
for name,param in model.named_parameters():
    if param.requires_grad:
        print("Layer ", index)
        print(name,param.data)
        print(name,torch.max(param.data))
        index += 1
```

```python
import torch
import torch.nn as nn
from collections import OrderedDict

data_in = torch.tensor([[0, 0],[1, 0],[0, 1],[1, 1]], dtype=torch.float32)
data_out = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]], dtype=torch.float32)

print("data in = ", data_in)
print("data out = ", data_out)

model = nn.Sequential(OrderedDict([
    ('hidden', nn.Linear(2,6)),
    ('tanh', nn.ReLU()),
    ('output', nn.Linear(6,2)),
    ('sigmoid', nn.Sigmoid())
]))

print("model = ", model)

criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.4)

for epoch in range(1000):
    output = model(data_in)
    loss = criterion(output, data_out)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 100 == 0:
        print(f'Epoch {epoch+1}/{1000} with loss =  {loss.item():.4f}')

#print(output)
predicted_list = torch.round(output)
#print(predicted_list == data_out)
correct = (predicted_list == data_out).sum()
accuracy = correct / (len(data_out) * 2)
print("Accuracy is = ", accuracy.item())

if accuracy == 1.0:
    start_time = datetime.strptime(start_time.strftime("%H:%M:%S:%f"),"%H:%M:%S:%f")
    stop_time = datetime.strptime(stop_time.strftime("%H:%M:%S:%f"),"%H:%M:%S:%f")
    print("Total time = ", stop_time - start_time)

print("Weights:")
index:int = 0
for name,param in model.named_parameters():
    if param.requires_grad:
        print("Layer ", index)
        print(name,param.data)
        print(name,torch.max(param.data))
        index += 1
```
